```python
import matplotlib.pyplot as plt
import numpy as np

xekseni = np.array([0, 7]) #x ekseninin cetvelleme ölçüsünü oluşturduk
yekseni = np.array([0, 250]) #y ekseninin cetvelleme ölçüsünü oluşturduk

plt.plot(xekseni, yekseni)
plt.show()
```

Bu plot()fonksiyon, bir diyagramda noktalar (işaretleyiciler) çizmek için kullanılır.

Varsayılan olarak, bu plot()fonksiyon noktadan noktaya bir çizgi çizer.

Bu fonksiyon, diyagramdaki noktaları belirtmek için parametreler alır.

Parametre 1, x ekseni üzerindeki noktaları içeren bir dizidir .

Parametre 2, y ekseni üzerindeki noktaları içeren bir dizidir .

(1, 3) noktasından (8, 10) noktasına bir çizgi çizmek istiyorsak, [1, 8] ve [3, 10] olmak üzere iki diziyi çizim fonksiyonuna iletmemiz gerekir.

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.title("Matplotlib demo")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#Çizgisiz Çizim
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
#Çoklu Noktalar
xpoints = np.array([2022, 2023, 2024, 2025])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

# **İşaretleyici Boyutu**

İşaretleyicilerin boyutunu ayarlamak için anahtar kelime bağımsız değişkenini markersizeveya daha kısa sürümünü kullanabilirsiniz

```python
#İşaretleyicilerin boyutunu 20 olarak ayarlayın:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 5, 10])

plt.plot(ypoints, marker = '*', ms = 20)
plt.show()
```

```python
#İşaretleyicilerin kenar rengini ayarlamak için anahtar kelime bağımsız
#değişkenini markeredgecolorveya daha kısa olanını kullanabilirsiniz :mec
#Kenar rengini kırmızı olarak ayarlayın:

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
#İşaretleyicilerin kenarı içindeki rengi ayarlamak için anahtar kelime bağımsız değişkenini markerfacecolorveya daha kısa olanını kullanabilirsiniz :mfc
#İç rengini kırmızı olarak ayarlayın:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 10, mfc = 'r')
plt.show()
```

```python
#İşaretleyicinin tamamını renklendirmek için hem `the` mechem de `then` argümanlarını kullanın :mfc
#Hem kenarın hem de yüzeyin rengini kırmızıya ayarlayın:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 10, mec = 'r', mfc = 'r')
plt.show()
```

```python
#Ayrıca onaltılık renk değerlerini de kullanabilirsiniz :
#Her noktayı güzel bir yeşil renkle işaretleyin:

ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
```

```python
#Veya desteklenen 140 renk adından herhangi biri .
#https://www.w3schools.com/colors/colors_names.asp

ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = '*', ms = 20, mec = 'lime', mfc = 'lime')
```

# **Matplotlib Çizgisi**

Çizilen çizginin stilini değiştirmek için anahtar kelime bağımsız değişkenini linestyleveya daha kısa olanını kullanabilirsiniz :ls

Çizgi stili daha kısa bir sözdizimiyle de yazılabilir:

linestyleşu şekilde yazılabilir ls: .

dottedşu şekilde yazılabilir :: .

dashedşu şekilde yazılabilir --: .

```python
plt.plot(ypoints, ls = '--',color = 'r')
```

# **Çizgi Genişliği**

Çizginin genişliğini değiştirmek için anahtar kelime argümanını linewidthveya daha kısa olanı kullanabilirsiniz .lw

Değer, ondalık sayı türünde ve puan cinsindendir:



```python
#20,5 punto genişliğinde bir çizgiyle çizim yapın:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5')
plt.show()
```
