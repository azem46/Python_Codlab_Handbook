**Python Data Type**

```python
#Text Type:	str

a= "Mehmet"
```

```python
#Numeric Types:	int, float, complex
a= 2
b= 2.5
c= 1j
print(type(a))
print(type(b))
print(type(c))
```

```python
#Sequence Types:	list, tuple, range
x = ["apple", "banana", "cherry"]
print(type(x))
x = ("apple", "banana", "cherry")
print(type(x))
x = range(6)
print(type(x))
```

```python
#Mapping Type:	dict
x = {"name" : "John", "age" : 36}
print(type(x))
```

```python
#Set Types:	set, frozenset
x = {"apple", "banana", "cherry"}
print(type(x))
x = frozenset({"apple", "banana", "cherry"})
print(type(x))
```

```python
#Boolean Type:	bool
x = True
print(type(x))
```

```python
#None Type:	NoneType
x = None
print(type(x))
```
