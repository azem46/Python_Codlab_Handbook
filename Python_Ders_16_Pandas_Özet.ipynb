# **Dizi nedir?**

# **Etiketler**
Başka bir şey belirtilmediği takdirde, değerler indeks numaralarıyla etiketlenir. Birinci değerin indeksi 0, ikinci değerin indeksi 1 vb. şeklindedir.

Bu etiket, belirtilen bir değere erişmek için kullanılabilir.

Etiketler Oluştur

# Boş Hücreler

Boş hücreler, verileri analiz ederken yanlış sonuçlara yol açabilir.

Satırları Kaldır
Boş hücrelerle başa çıkmanın bir yolu, boş hücre içeren satırları silmektir.

Bu genellikle sorun teşkil etmez, çünkü veri kümeleri çok büyük olabilir ve birkaç satırın kaldırılması sonuç üzerinde büyük bir etki yaratmaz.

```python
#Boş hücre içermeyen yeni bir Veri Çerçevesi döndürün:

import pandas as pd

df = pd.read_csv('data.csv')

new_df = df.dropna()

print(new_df.to_string())
```

Not: Varsayılan olarak, bu yöntem yenidropna() bir DataFrame döndürür ve orijinalini değiştirmez.

Orijinal DataFrame'i değiştirmek istiyorsanız şu inplace = Trueargümanı kullanın:

```python
#Boş (NULL) değer içeren tüm satırları silin:

import pandas as pd

df = pd.read_csv('data.csv')

df.dropna(inplace = True)

print(df.to_string())
```

Not: Bu işlem dropna(inplace = True)yeni bir DataFrame döndürmeyecek, ancak orijinal DataFrame'den NULL değerleri içeren tüm satırları kaldıracaktır.

# Boş Değerleri Değiştir
Boş hücrelerle başa çıkmanın bir başka yolu da, bunun yerine yeni bir değer eklemektir .

Bu sayede, sadece birkaç boş hücre yüzünden tüm satırları silmek zorunda kalmazsınız.

Bu fillna()yöntem, boş hücreleri bir değerle değiştirmemizi sağlar:

```python
#Boş değerleri 130 sayısıyla değiştirin:

import pandas as pd

df = pd.read_csv('data.csv')

df.fillna(130, inplace = True)
```

# Yalnızca Belirtilen Sütunlar İçin Değiştir
Yukarıdaki örnek, tüm Veri Çerçevesindeki boş hücrelerin tamamını değiştirir.

Yalnızca bir sütundaki boş değerleri değiştirmek için, DataFrame'in sütun adını belirtin:

```python
#"Kalori" sütunundaki boş (NULL) değerleri 130 sayısıyla değiştirin:

import pandas as pd

df = pd.read_csv('data.csv')

df.fillna({"Calories": 130}, inplace=True)
```

# Ortalama, Medyan veya Mod Kullanarak Değiştirin
Boş hücreleri doldurmanın yaygın bir yolu, sütunun ortalama, medyan veya mod değerini hesaplamaktır.

Pandas, belirtilen bir sütun için ilgili değerleri hesaplamak üzere mean() median()`and` yöntemlerini kullanır :mode()

```python
#Ortalamayı hesaplayın ve boş kalan değerleri bu ortalama ile değiştirin:

import pandas as pd

df = pd.read_csv('data.csv')

x = df["Weight"].mean()

print(df.fillna({"Weight": x}, inplace=True))
```

Medyan = tüm değerler artan sırada sıralandıktan sonra ortadaki değer.

```python
#Modu hesaplayın ve boş olan değerleri bu mod ile değiştirin:

import pandas as pd

df = pd.read_csv('data.csv')

x = df["Weight"].mode()[0]

print(df.fillna({"Weight": x}, inplace=True))

```

# **Sonuç Açıklaması**
Bu yöntemin sonucu, corr()iki sütun arasındaki ilişkinin ne kadar iyi olduğunu gösteren birçok sayı içeren bir tablodur.

Sayı -1 ile 1 arasında değişir.

1, bire bir ilişki (mükemmel korelasyon) olduğu anlamına gelir ve bu veri kümesi için, birinci sütundaki bir değer her yükseldiğinde, diğer sütundaki değer de yükselmiştir.

0,9 da iyi bir ilişki oranıdır ve bir değeri artırırsanız, diğer değer de muhtemelen artacaktır.

-0.9 değeri de 0.9 kadar iyi bir ilişki olurdu, ancak bir değeri artırırsanız, diğeri muhtemelen düşer.

0,2 iyi bir ilişki olmadığını gösterir; yani bir değerin yükselmesi diğerinin de yükseleceği anlamına gelmez.

İyi bir korelasyon nedir? Kullanım amacına bağlıdır, ancak iyi bir korelasyon olarak adlandırmak için en az 0.6(veya ) olması gerektiğini söylemek güvenli olur diye düşünüyorum .-0.6

"Süre" ve "Süre" sütunlarının aynı sayıyı aldığını görüyoruz 1.000000, bu da mantıklı; her sütunun kendisiyle mükemmel bir ilişkisi var.

İyi Korelasyon:
"Süre" ve "Kalori" 0.922721arasında çok iyi bir korelasyon var ve şunu tahmin edebiliriz: Ne kadar uzun süre egzersiz yaparsanız, o kadar çok kalori yakarsınız ve bunun tersi de geçerlidir: eğer çok kalori yaktıysanız, muhtemelen uzun bir egzersiz yapmışsınızdır.

Kötü Korelasyon:
"Süre" ve "Maksimum Nabız" 0.009403arasında çok zayıf bir korelasyon var; yani sadece egzersizin süresine bakarak maksimum nabzı tahmin edemeyiz ve bunun tersi de geçerli.
