# **Python Döngüleri**

*Python'da iki temel döngü komutu vardır:*

while döngüleri

for döngüleri

```python
#while Döngüsü
#Not: i'yi artırmayı unutmayın, aksi takdirde döngü sonsuza kadar devam eder.
while i < 6:
  print(i)
  i += 1
```

```python
#i değeri 3 olduğunda döngüden çık:
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
#i değeri 3 ise bir sonraki yinelemeye geçin:
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

```python
#Koşul yanlış olduğunda bir mesaj yazdırın:
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```
