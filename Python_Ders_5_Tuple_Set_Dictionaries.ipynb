# **Tuple**

Demetler (tuples), birden fazla öğeyi tek bir değişkende saklamak için kullanılır.

Python'da veri koleksiyonlarını depolamak için kullanılan 4 yerleşik veri türünden biri olan Tuple, diğer 3'ü List , Set ve Dictionary'dir ve her birinin farklı özellikleri ve kullanım alanları vardır.

Bir demet (tuple), sıralı ve değiştirilemez bir koleksiyondur .

Demetler yuvarlak parantez içinde yazılır.

```python
#Bir demet (tuple) oluşturmak İçin:
thistuple = ("elma", "muz", "portakal","muz")
print(thistuple)
```

```python
#Tuple uzunluğunu hesaplamak İçin:
print(len(thistuple))
```

```python
#Tuple öğeleri herhangi bir veri türünde olabilir:
tuple1 = ("togg", "byd", "tesla")
tuple2 = (1, 5, 7, 9, 3)
tuple3 = (True, False, False)
```
