# **Matplotlib Etiketleri ve Başlığı**

Pyplot ile x ve y eksenleri için etiket belirlemek üzere  xlabel()ve ylabel()fonksiyonlarını kullanabilirsiniz .

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 150, 270, 280, 290, 170, 310, 320, 330])

plt.plot(x, y,
         color = "red",
         linewidth = "2.5",
         linestyle = "--",
         )

plt.title("Sports Watch Data") #Başlık
plt.xlabel("Average Pulse")    #x yatay eksen başlığı
plt.ylabel("Calorie Burnage")  #y yatay eksen bağlığı

plt.show()
```

# **Başlık ve Etiketler için Yazı Tipi Özelliklerini Ayarlayın**


Başlık ve etiketler için yazı tipi özelliklerini ayarlamak üzere , , ve fontdictparametrelerini kullanabilirsiniz .xlabel()ylabel()title()

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

# **Başlığı Konumlandırın**


Başlığı konumlandırmak için bu locparametreyi kullanabilirsiniz .title()

Geçerli değerler şunlardır: 'sol', 'sağ' ve 'merkez'. Varsayılan değer 'merkez'dir.

```python
#Başlığı sola hizalayın:

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

# **Grafiğe Izgara Çizgileri Ekle**

grid()Pyplot ile, grafiğe kılavuz çizgileri eklemek için fonksiyonunu kullanabilirsiniz .

```python
#Grafiğe kılavuz çizgileri ekleyin:

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 200, 310, 320, 230])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

**Hangi kılavuz çizgilerinin görüntüleneceğini belirtin.**

axisFonksiyon içindeki parametreyi kullanarak grid()hangi kılavuz çizgilerinin görüntüleneceğini belirtebilirsiniz.

Geçerli değerler şunlardır: 'x', 'y' ve 'both'. Varsayılan değer 'both'tur.

```python
#Yalnızca x ekseni için kılavuz çizgilerini göster:

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```

```python
#Yalnızca y ekseni için kılavuz çizgilerini göster:

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'y')

plt.show()
```

**Izgara için Çizgi Özelliklerini Ayarlayın**

Izgaranın çizgi özelliklerini şu şekilde de ayarlayabilirsiniz: grid(color = ' color ', linestyle = ' linestyle ', linewidth = number )

```python
#Izgaranın çizgi özelliklerini ayarlayın:

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

# **Matplotlib Alt Grafiği**

Birden Çok Grafiği Görüntüle
Bu fonksiyon sayesinde subplot()tek bir şekilde birden fazla grafiği çizebilirsiniz:

```python
#İki grafik çizin:
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

# **subplot() fonksiyonu**

Bu subplot()fonksiyon, şeklin düzenini tanımlayan üç argüman alır.

Düzen, birinci ve ikinci argümanlarla temsil edilen satır ve sütunlar şeklinde organize edilmiştir.

Üçüncü argüman, mevcut grafiğin indeksini temsil eder.(satır,sütun,index)

```python
#İki grafiği üst üste çizin:
#Bir şekil üzerinde istediğiniz kadar grafik çizebilirsiniz, sadece satır ve sütun sayısını ve grafiğin indeksini belirtmeniz yeterlidir.

import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```
