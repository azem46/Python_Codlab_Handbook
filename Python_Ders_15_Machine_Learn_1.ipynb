# **Standart sapma nedir?**

Standart sapma, değerlerin ne kadar yayılmış olduğunu açıklayan bir sayıdır.

Düşük standart sapma, sayıların çoğunun ortalama değere yakın olduğu anlamına gelir.

Yüksek standart sapma, değerlerin daha geniş bir aralığa yayılmış olduğu anlamına gelir.

Örnek: Bu sefer 7 aracın hızını kaydettik:

speed = [86,87,88,86,87,85,86]

Standart sapma şöyledir:

0.9

Bu, değerlerin çoğunun ortalama değer olan 86,4'ten 0,9'luk bir aralık içinde olduğu anlamına gelir.

Şimdi de daha geniş bir aralıktaki sayılarla aynı işlemi yapalım:

speed = [32,111,138,28,59,77,97]

Standart sapma şöyledir:

37.85

Bu, değerlerin çoğunun ortalama değer olan 77,4'ten 37,85 aralığında olduğu anlamına gelir.

Gördüğünüz gibi, daha yüksek bir standart sapma, değerlerin daha geniş bir aralığa yayıldığını gösterir.

NumPy modülü, standart sapmayı hesaplamak için bir yönteme sahiptir:

```python
#NumPy std()yöntemini kullanarak standart sapmayı bulun:

import numpy

speed = [86,87,88,86,87,85,86]

x = numpy.std(speed)

print(x)
```

```python
import numpy

speed = [32,111,138,28,59,77,97]

x = numpy.std(speed)

print(x)
```

# **Varyans**


Varyans, değerlerin ne kadar yayılmış olduğunu gösteren bir başka sayıdır.

Aslında, varyansın karekökünü alırsanız standart sapmayı elde edersiniz!

Ya da tam tersi, standart sapmayı kendisiyle çarparsanız varyansı elde edersiniz!

Varyansı hesaplamak için aşağıdaki adımları izlemeniz gerekir:

1. Ortalamayı bulun:

(32+111+138+28+59+77+97) / 7 = 77.4

2. Her değer için, ortalamadan farkını bulun:

 32 - 77.4 = -45.4

111 - 77.4 =  33.6

138 - 77.4 =  60.6

 28 - 77.4 = -49.4

 59 - 77.4 = -18.4

 77 - 77.4 = - 0.4

 97 - 77.4 =  19.6

3. Her fark için karesini bulun:

(-45.4)2 = 2061.16

 (33.6)2 = 1128.96

 (60.6)2 = 3672.36

(-49.4)2 = 2440.36

(-18.4)2 =  338.56

(- 0.4)2 =    0.16

 (19.6)2 =  384.16

4. Varyans, bu kare farklarının ortalama sayısıdır:

(2061.16+1128.96+3672.36+2440.36+338.56+0.16+384.16) / 7 = 1432.2

```python
#var()Varyansı bulmak için NumPy yöntemini kullanın :

import numpy

speed = [32,111,138,28,59,77,97]

x = numpy.var(speed)

print(x)
```

```python
#NumPy std()yöntemini kullanarak standart sapmayı bulun:

import numpy

speed = [32,111,138,28,59,77,97]

x = numpy.std(speed)

print(x)
```

Standart sapma ve varyans, makine öğreniminde sıklıkla kullanılan terimlerdir; bu nedenle bunların nasıl elde edildiğini ve arkalarındaki kavramı anlamak önemlidir.

#Yüzdelikler nedir?

İstatistik biliminde yüzdelik dilimler, değerlerin belirli bir yüzdesinin hangi değerden daha düşük olduğunu gösteren bir sayı vermek için kullanılır.

Örnek: Diyelim ki, bir sokakta yaşayan her kişinin yaşını içeren bir dizimiz var.

75. yüzdelik dilim kaçtır? Cevap 43'tür; bu da insanların %75'inin 43 yaşında veya daha genç olduğu anlamına gelir.

NumPy modülü, belirtilen yüzdelik dilimi bulmak için bir yönteme sahiptir:

```python
#percentile()Yüzdelik dilimleri bulmak için NumPy yöntemini kullanın :

import numpy

ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

x = numpy.percentile(ages, 75)

print(x)
```

```python
#Nüfusun %90'ının daha genç olduğu yaş kaçtır?

import numpy

ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

x = numpy.percentile(ages, 90)

print(x)
```

# **Regresyon**

Değişkenler arasındaki ilişkiyi bulmaya çalışırken regresyon terimi kullanılır.

Makine öğreniminde ve istatistiksel modellemede, bu ilişki gelecekteki olayların sonucunu tahmin etmek için kullanılır.

Doğrusal Regresyon
Doğrusal regresyon, veri noktaları arasındaki ilişkiyi kullanarak tüm bu noktalardan geçen düz bir çizgi çizer.

Bu çizgi, gelecekteki değerleri tahmin etmek için kullanılabilir.

Nasıl Çalışır?
Python, veri noktaları arasındaki ilişkiyi bulmak ve doğrusal regresyon doğrusu çizmek için yöntemlere sahiptir. Matematiksel formüllerle uğraşmak yerine, bu yöntemleri nasıl kullanacağınızı göstereceğiz.

Aşağıdaki örnekte, x ekseni yaşı, y ekseni ise hızı temsil etmektedir. Gişeden geçerken 13 aracın yaşını ve hızını kaydettik. Topladığımız verilerin doğrusal regresyonda kullanılıp kullanılamayacağına bakalım:

```python
#Öncelikle bir dağılım grafiği çizerek başlayın:

import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.show()
```

```python
#Doğrusal regresyon modelini içe aktarın scipyve doğru çizgisini çizin:

import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

# **R, İlişki anlamına gelir.**

X ekseni değerleri ile Y ekseni değerleri arasındaki ilişkinin nasıl olduğunu bilmek önemlidir; eğer bir ilişki yoksa, doğrusal regresyon herhangi bir şeyi tahmin etmek için kullanılamaz.

Bu ilişkiye -korelasyon katsayısına- denir r.

Değer r-1 ile 1 arasında değişir; 0 hiçbir ilişki olmadığını, 1 (ve -1) ise %100 ilişki olduğunu gösterir.

Python ve Scipy modülü bu değeri sizin için hesaplayacaktır, yapmanız gereken tek şey x ve y değerlerini ona iletmektir.

```python
#Verilerim doğrusal regresyona ne kadar iyi uyuyor?

from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

print(r)
```

Not: -0.76 sonucu, mükemmel olmasa da bir ilişki olduğunu göstermektedir; bu da gelecekteki tahminlerde doğrusal regresyon kullanabileceğimizi işaret etmektedir.

#Polinom Regresyonu

Veri noktalarınız doğrusal regresyona (tüm veri noktalarından geçen düz bir çizgi) açıkça uymuyorsa, polinom regresyonu ideal olabilir.

Polinom regresyonu, doğrusal regresyon gibi, veri noktaları arasından bir çizgi çizmenin en iyi yolunu bulmak için x ve y değişkenleri arasındaki ilişkiyi kullanır.

Nasıl Çalışır?
Python, veri noktaları arasındaki ilişkiyi bulmak ve polinom regresyon doğrusu çizmek için yöntemlere sahiptir. Matematiksel formüllerle uğraşmak yerine, bu yöntemleri nasıl kullanacağınızı göstereceğiz.

Aşağıdaki örnekte, belirli bir gişeden geçerken 18 aracı kaydettik.

Aracın hızını ve geçişin gerçekleştiği saati (günün hangi saatinde olduğunu) kaydettik.

# **R-Kare**

X ve Y eksenlerinin değerleri arasındaki ilişkinin ne kadar iyi olduğunu bilmek önemlidir; eğer bir ilişki yoksa, polinom regresyonu herhangi bir şeyi tahmin etmek için kullanılamaz.

Bu ilişki, r-kare adı verilen bir değerle ölçülür.

R-kare değeri 0 ile 1 arasında değişir; 0 hiçbir ilişki olmadığını, 1 ise %100 oranında ilişki olduğunu gösterir.

Python ve Sklearn modülü bu değeri sizin için hesaplayacaktır, yapmanız gereken tek şey x ve y dizilerini ona iletmektir:

```python
#Verilerim polinom regresyonuna ne kadar iyi uyuyor?

import numpy
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

print(r2_score(y, mymodel(x)))
```

Not: Elde edilen 0,94 sonucu, çok iyi bir ilişki olduğunu ve gelecekteki tahminlerde polinom regresyonunu kullanabileceğimizi göstermektedir.
