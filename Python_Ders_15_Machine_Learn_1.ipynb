**Veri Tipleri**

Verileri analiz etmek için, hangi tür verilerle uğraştığımızı bilmek önemlidir.

Veri türlerini üç ana kategoriye ayırabiliriz:

Sayısal
Kategorik
Sıralı

Sayısal veriler sayılardır ve iki sayısal kategoriye ayrılabilirler:

**Ayrık Veri**
- tamsayılarla sınırlı olan sayılabilir veriler. Örnek: Geçen araba sayısı.

**Sürekli Veri**
- herhangi bir sayı olabilen ölçülen veri. Örnek: Bir ürünün fiyatı veya bir ürünün boyutu.

**Kategorik veriler**, birbirleriyle karşılaştırılamayan değerlerdir. Örnek: bir renk değeri veya herhangi bir evet/hayır değeri.

**Sıralı veriler**, kategorik verilere benzer, ancak birbirleriyle karşılaştırılabilirler. Örnek: Okul notları; A, B'den daha iyidir ve benzeri.

**Veri kaynağınızın veri türünü bilerek**, analiz yaparken hangi tekniği kullanacağınızı da anlayabilirsiniz.
