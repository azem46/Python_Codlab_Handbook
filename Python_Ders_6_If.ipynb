# **Python If Deyimi**

Python, matematikteki olağan mantıksal koşulları destekler:

Eşittir: **a == b**

Eşit Değil: **a != b**

Küçüktür: **a < b**

Küçük veya eşit: **a <= b**

Büyüktür: **a > b**

Büyük veya eşit: **a >= b**

Bu koşullar çeşitli şekillerde kullanılabilir, en yaygın kullanımları ise "if" ifadelerinde ve döngülerdedir.

`'Eğer ifadesi'`, `if` anahtar kelimesi kullanılarak yazılır .

```python
#Eğer koşulu:
a = 33
b = 200
if b > a:
  print("b is greater than a")
```
