# **Python Iterators**

Listeler, demetler, sözlükler ve kümeler, yinelenebilir nesnelerdir. Bunlar, içlerinden yineleyici elde edebileceğiniz yinelenebilir kaplardır .

Bu nesnelerin hepsinin iter()yineleyici elde etmek için kullanılan bir yöntemi vardır

```python
#Bir demetten yineleyici döndürün ve her değeri yazdırın:

mytuple = ("apple", "banana", "cherry")
myit = iter(mytuple)

print(next(myit))
print(next(myit))
print(next(myit))
```

```python
#Dizeler de karakter dizisi içeren, yinelenebilir nesnelerdir:

mystr = "banana"
myit = iter(mystr)

print(next(myit))
print(next(myit))
print(next(myit))
print(next(myit))
print(next(myit))
print(next(myit))
```

```python
#Bir demetin değerleri üzerinde yineleme yapın:

mytuple = ("apple", "banana", "cherry")

for x in mytuple:
  print(x)
```

```python
#Bir dizenin karakterleri üzerinde yineleme yapın:
#Döngü foraslında bir yineleyici nesnesi oluşturur ve next() her döngü için yöntemi yürütür.

mystr = "banana"

for x in mystr:
  print(x)
```

# **Python Dates**

```python
#Tarih ve saat modülünü içe aktarın ve geçerli tarihi görüntüleyin:
import datetime

x = datetime.datetime.now()
print(x)
```

```python
#Yılı ve haftanın gününün adını girin:

import datetime

x = datetime.datetime.now()

print(x.year)
print(x.strftime("%A"))
```

```python
#Bu datetime()sınıf, tarih oluşturmak için üç parametre gerektirir: yıl, ay, gün.
#Bir tarih nesnesi oluşturun:

import datetime

x = datetime.datetime(2020, 5, 17)

print(x)
```

**strftime() Metodu**

```python
#Bu yönteme adı verilir strftime()ve formatdöndürülecek dizenin biçimini belirtmek için tek bir parametre alır:
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%B"))
```

```python
#metodunu kullanarak tarihi ay/gün/yıl (MM/DD/YY) biçiminde (06/01/18 gibi) biçimlendirir ve ekrana yazdırır.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%D"))
```

```python
#yöntemini kullanarak yılı dört haneli olarak (örneğin, 2018 için '2018' olarak) biçimlendirir ve ekrana yazdırır.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%Y"))
```

```python
# bu tarihin saatini 12 saat formatında (01-12 arası) biçimlendirir ve ekrana yazdırır. Örneğin, öğleden sonra saat 3 ise '03' olarak gösterir.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%I"))
```

```python
# bu tarihin AM/PM değerini (örneğin, 'AM' veya 'PM' olarak) biçimlendirir ve ekrana yazdırır.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%p"))
```

```python
#bu tarihin dakikasını (00-59 arası) biçimlendirir ve ekrana yazdırır.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%M"))
```

```python
#bu tarihin saniyesini (00-59 arası) biçimlendirir ve ekrana yazdırır.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%S"))
```

```python
# milisaniyeleri (altı basamaklı bir ondalık sayı olarak) biçimlendirir ve ekrana yazdırır.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%f"))
```

```python
#k UTC ofsetini (örneğin, '+0200' gibi) biçimlendirir ve ekrana yazdırır. Ancak, oluşturulan tarih nesnesi herhangi bir zaman dilimi
#bilgisi içermediği için bu durumda genellikle boş bir dize döndürecektir.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%z"))
```

```python
#yöntemini kullanarak zaman diliminin adını biçimlendirmeye çalışır ve ekrana yazdırır. Ancak, oluşturulan
#tarih nesnesi herhangi bir zaman dilimi bilgisi içermediği için bu durumda genellikle boş bir dize döndürür.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%Z"))
```

```python
#bu tarihin yılın kaçıncı günü olduğunu (örneğin, 1 Haziran 2018, yılın 152. günü olduğu için '152' olarak) biçimlendirir ve ekrana yazdırır.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%j"))
```

```python
# yılın hafta numarasını (Pazar haftanın ilk günü kabul edilerek,
#yılın ilk Pazar gününden önceki günler 0. hafta olarak sayılır) biçimlendirir ve ekrana yazdırır.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%U"))
```

```python
#yöntemini kullanarak yılın hafta numarasını (Pazartesi haftanın ilk günü kabul edilerek,
#yılın ilk Pazartesi gününden önceki günler 0. hafta olarak sayılır) biçimlendirir ve ekrana yazdırır.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%W"))
```

```python
#bu tarihin yerel ayar için uygun tarih ve saat gösterimini (örneğin, 'Fri Jun 1 00:00:00 2018' gibi) biçimlendirir ve ekrana yazdırır.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%c"))
```

```python
#bu tarihin yüzyılını (örneğin, 2018 için '20' olarak) biçimlendirir ve ekrana yazdırır.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%C"))
```

```python
#bu tarihin yerel ayar için uygun tarih gösterimini biçimlendirir ve ekrana yazdırır.
#Bu format genellikle ay/gün/yıl (MM/DD/YY) şeklinde olur, ancak kullanıcının yerel ayarlarına göre değişebilir.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%x"))
```

```python
#bu tarihin yerel ayar için uygun saat gösterimini biçimlendirir ve ekrana yazdırır.
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%X"))
```

```python
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%%"))
```
